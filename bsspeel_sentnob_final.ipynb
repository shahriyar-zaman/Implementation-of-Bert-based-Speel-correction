```python
from bnlp import CleanText
```

```python
clean_text = CleanText(
   fix_unicode=True,
   unicode_norm=True,
   unicode_norm_form="NFKC",
   remove_url=False,
   remove_email=False,
   remove_emoji=False,
   remove_number=False,
   remove_digits=False,
   remove_punct=False,
   replace_with_url="<URL>",
   replace_with_email="<EMAIL>",
   replace_with_number="<NUMBER>",
   replace_with_digit="<DIGIT>",
   replace_with_punct = "<PUNC>"
)

df_trainset['Data'] = df_trainset['Data'].apply(clean_text)
df_testset['Data'] = df_testset['Data'].apply(clean_text)
```

```python
##After cleaning, Longest Sentence words in trainSet
longest_string = max(df_trainset["Data"], key=len)
print(len(longest_string))
```

```python
##After cleaning, Sentence words in testSet
longest_string = max(df_testset["Data"], key=len)
print(len(longest_string))
```

##Training Model

```python
# from sklearn.model_selection import train_test_split
train = df_trainset['Data'].values.tolist()
test = df_testset['Data'].values.tolist()
train_cluster_id = df_trainset['Label'].tolist()
test_cluster_id = df_testset['Label'].tolist()
```

```python
X_train = train
y_train = train_cluster_id
X_test = test
y_test = test_cluster_id
```

```python
pip install ktrain
```

```python
import ktrain
from ktrain import text
MODEL_NAME = 'csebuetnlp/banglabert'
t = text.Transformer(MODEL_NAME, maxlen=500)
```

```python
y_train = [str(x) for x in train_cluster_id]
y_test = [str(x) for x in test_cluster_id]
```

```python
trn = t.preprocess_train(X_train, y_train)
val = t.preprocess_test(X_test, y_test)
```

```python
model = t.get_classifier()

learner = ktrain.get_learner(model, train_data=trn, val_data=val, batch_size=10)

learner.fit_onecycle(1e-4, 5)
```

```python
learner.validate(class_names=t.get_classes())
```
